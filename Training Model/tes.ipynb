```python
import numpy as np
import pandas as pd
import joblib
import json
from pathlib import Path
```

```python
ARTIFACTS_DIR = Path("artifacts")

with open(ARTIFACTS_DIR / "feature_columns.json", "r") as f:
    feature_columns = json.load(f)

model_dt = joblib.load(ARTIFACTS_DIR / "job_predictor_dt.pth")

print("Artifacts loaded!")
print("Total fitur:", len(feature_columns))
```

```python
def get_user_input(feature_columns):
    print("=== Masukkan nilai skill kamu (0-10) ===\n")
    data = {}
    
    for ft in feature_columns:
        while True:
            try:
                v = float(input(f"{ft}: "))
                if 0 <= v <= 10:
                    data[ft] = v
                    break
                else:
                    print("Masukkan angka 0–10!")
            except:
                print("Input tidak valid!")
                
    return pd.DataFrame([data])

user_df = get_user_input(feature_columns)
user_df
```

```python
pred_dt = model_dt.predict(user_df)[0]

print("Job terbaik untuk skill kamu:", pred_dt)
```
